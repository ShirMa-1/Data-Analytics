## Задание 1

Даны 2 переменных, в которых хранятся строки произвольной длины: *phrase_1* и *phrase_2*.  
Напишите код, который проверяет какая из этих строк длиннее.

Примеры работы программы:  
1.  

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

```

Результат:  
`Фраза 1 длиннее фразы 2`
2.     

```
phrase_1 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:  
`Фраза 2 длиннее фразы 1`
3.     

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:  
`Фразы равной длины`

````python
phrase_1 = input("Фраза 1: ")
phrase_2 = input("Фраза 2: ")

if len(phrase_2) < len(phrase_1):
    print("Фраза 1 длиннее фразы 2")
elif len(phrase_1) < len(phrase_2):
    print("Фраза 2 длиннее фразы 1")
else:
    print("Фразы равной длины")
````

## Задание 2

Дана переменная, в которой хранится четырехзначное число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы:
1.  

```
year = 2020
```
Результат:  
`Високосный год` 

2.  

```
year = 2019
```
Результат:  
`Обычный год` 

````python
year = int(input("Год: "))

# Если год НЕ делится на 4, значит он ОБЫЧНЫЙ
# Если год НЕ делится на 100, значит он ВИСОКОСНЫЙ
# Если год делится на 100, но делится на 400, то он ВИСОКОСНЫЙ
        
if year%4==0 and year%100!=0 or year%400==0:
    print("Високосный год")
else:
    print("Обычный год")
````

## Задание 3

Необходимо написать программу, которая будет запрашивать у пользователя месяц и дату рождения и выводить соответствующий знак зодиака.

Пример работы программы:  
1.  

```
Введите день:
30

Введите месяц:
Август
```
Результат:  
`Ваш знак зодиака: Дева` 

2.  

```
Введите день:
29

Введите месяц:
Октябрь
```
Результат:  
`Ваш знак зодиака: Скорпион` 


````python
day = int(input("Введите день: \n"))
month = input("Введите месяц (в именительном патеже с большой буквы): \n")

# Водолей (21 января - 20 февраля)
# Рыбы (21 февраля - 20 марта)
# Овен (21 марта - 20 апреля)
# Телец (21 апреля - 20 мая)
# Близнецы (21 мая - 21 июня)
# Рак (22 июня - 22 июля)
# Лев (23 июля - 23 августа)
# Дева (24 августа - 23 сентября)
# Весы (24 сентября - 23 октября)
# Скорпион (24 октября - 22 ноября)
# Стрелец (23 ноября - 21 декабря)
# Козерог (22 декабря - 20 января)

if month == "Январь" and 21 <= day <= 31 or month == "Февраль" and 0 < day <= 20:
    print("Ваш знак зодиака: Водолей")  
elif month == "Февраль" and  21 <= day <= 29 or month == "Март" and 0 < day <= 20:
    print("Ваш знак зодиака: Рыбы")  
elif month == "Март" and 21 <= day <= 31 or month == "Апрель" and 0 < day <= 20:
    print("Ваш знак зодиака: Овен")  
elif month == "Апрель" and 21 <= day <= 30 or month == "Май" and 0 < day <= 20:
    print("Ваш знак зодиака: Телец")  
elif month == "Май" and 21 <= day <= 31 or month == "Июнь" and 0 < day <= 21:
    print("Ваш знак зодиака: Близнецы")  
elif month == "Июнь" and 22 <= day <= 30 or month == "Июль" and 0 < day <= 22:
    print("Ваш знак зодиака: Рак")  
elif month == "Июль" and 23 <= day <= 31 or month == "Август" and 0 < day <= 23:
    print("Ваш знак зодиака: Лев")  
elif month == "Август" and 24 <= day <= 31 or month == "Сентябрь" and 0 < day <= 23:
    print("Ваш знак зодиака: Дева")  
elif month == "Сентябрь" and 24 <= day <= 30 or month == "Октябрь" and 0 < day <= 23:
    print("Ваш знак зодиака: Весы")  
elif month == "Октябрь" and 24 <= day <= 31 or month == "Ноябрь" and 0 < day <= 22:
    print("Ваш знак зодиака: Скорпион")  
elif month == "Ноябрь" and 23 <= day <= 30 or month == "Декабрь" and 0 < day <= 21:
    print("Ваш знак зодиака: Стрелец")  
elif month == "Декабрь" and 21 <= day <= 31 or month == "Январь" and 0 < day <= 20:
    print("Ваш знак зодиака: Козерог")   
else:
    print("Ошибка в вводе данных")
````

## Задание 4

Вам нужно написать программу для подбора упаковок по размерам товара. Размеры хранятся в переменных (в сантиметрах):

```
width = 10
length = 205
height = 5
```

Используйте следующие правила:
- если каждое из трех измерений менее 15 сантиметров, то выведите на экран "Коробка №1";
- если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите "Коробка №2";
- если длина товара больше 2 метров, то выводите "Упаковка для лыж";
- во всех остальных случаях выводите "Стандартная коробка №3".


````python
width = int(input("Ширина(см) = "))
length = int(input("Длина(см) = "))
height = int(input("Высота(см) = "))

if width < 15 and length < 15 and height < 15:
    print("Коробка №1")
elif 15 < width < 50 or 15 < length < 50 or  15 < height < 50:
    print("Коробка №2")
elif length > 200:
    print("Упаковка для лыж")
else:
    print("Стандартная коробка №3")
````

## Задание 5

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.   

Примеры работы программы:  
1. ```number = 123456```

Результат:  
`Неасчастливый билет`
2. ```number = 123321```

Результат:  
`Счастливый билет`


````python
number = int(input("Номер билета: "))

rank_1 = number % 10
number = number // 10
rank_2 = number % 10
number = number // 10
rank_3 = number % 10
number = number // 10
rank_4 = number % 10
number = number // 10
rank_5 = number % 10
rank_6 = number // 10

if rank_1 + rank_2 + rank_3 == rank_4 + rank_5 + rank_6:
      print("Счастливый билет")
else:
    print("Несчастливый билет")
````

## Задание 6

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

Пример работы программы:
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:10
10
```
Результат:  
`Площадь круга: 314.16`

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

````python
figure = input("Фигура: \n")

if figure == "Круг":
    R = float(input("Введите радиус круга: \n"))
    S = 3.1416 * R * R
    flag = True
elif figure == "Треугольник":
    A = float(input("Введите длину стороны А: \n"))
    B = float(input("Введите длину стороны B: \n"))
    C = float(input("Введите длину стороны C: \n"))
    p = (A + B + C) / 2.0
    S = (p * (p - A) *(p - B) *(p - C)) ** 0.5
    flag = True
elif figure == "Прямоугольник":
    A = float(input("Введите длину стороны А: \n"))
    B = float(input("Введите длину стороны B: \n"))
    S = A * B
    flag = True
else:
    flag = False
    print("Не могу вычислить площать этой фигур")
if flag == True:
    print("Площадь " + figure.lower() + "а = ", round(S, 2))
````
